```python
merged_data = pd.read_csv('../data/merged_records_df.csv')
```

```python
mo_codes_all = merged_data['Mo Codes']
```

```python
import numpy as np
all_mo_codes = []
mo_codes_col = []
for i in mo_codes_all:
    mo_codes_col.append(i.split(" "))
    all_mo_codes = all_mo_codes + i.split(" ")
    
merged_data['Mo_Codes_List'] = mo_codes_col
unique_mo_codes = np.unique(all_mo_codes) #112 unique codes
```

```python
merged_data[['Description', 'Mo Codes', 'Crm Cd', 'Crm Cd 1', 'Crm Cd 2', 'Crm Cd 3',
       'Crm Cd 4', 'Crm Cd Desc', 'Prem Is Cd', 'Weapon Used Cd', 'Mo_Codes_List']]
```

```python
unique_mo_codes
```

```python
from collections import Counter
from nltk import ngrams
import re, string
bigtxt = merged_data['Description'][:2]
bigtxt = " ".join(bigtxt)
bigtxt = bigtxt.lower()
bigtxt ="".join(c for c in bigtxt if c not in ('!','.',':', ';', ','))
bigtxt = ' '.join(bigtxt.split())
ngram_counts = Counter(ngrams(bigtxt.lower().split(), 2))
ngram_counts.most_common(1)[0][0]
ngram_counts.most_common(1)[0][1]
```

```python
num_records = []
most_frequent_unigram = []
most_frequent_unigram_cnt = []
most_frequent_bigram = []
most_frequent_bigram_cnt = []
most_frequent_trigram = []
most_frequent_trigram_cnt = []
most_frequent_quadgram = []
most_frequent_quadgram_cnt = []
most_frequent_pentagram = []
most_frequent_pentagram_cnt = []

for code in unique_mo_codes:
    print("code:", code)
    sub_data = merged_data[merged_data['Mo_Codes_List'].apply(lambda x: str(code) in x)]['Description']
    num_records.append(len(sub_data))
    sub_data = [x for x in sub_data if str(x) != 'nan']

    if len(sub_data) > 0:
        bigtxt = " ".join(sub_data)
        bigtxt = bigtxt.lower()
        bigtxt ="".join(c for c in bigtxt if c not in ('!','.',':', ';', ','))
        bigtxt = ' '.join(bigtxt.split())
#         print("data :", bigtxt)
        for n in range(1, 6, 1):
            ngram_counts = Counter(ngrams(bigtxt.lower().split(), n))
            if n == 1:
                most_frequent_unigram.append(ngram_counts.most_common(1)[0][0])
                most_frequent_unigram_cnt.append(ngram_counts.most_common(1)[0][1])
            elif n == 2:
                most_frequent_bigram.append(ngram_counts.most_common(1)[0][0])
                most_frequent_bigram_cnt.append(ngram_counts.most_common(1)[0][1])
            elif n==3:
                most_frequent_trigram.append(ngram_counts.most_common(1)[0][0])
                most_frequent_trigram_cnt.append(ngram_counts.most_common(1)[0][1])
            elif n==4:
                most_frequent_quadgram.append(ngram_counts.most_common(1)[0][0])
                most_frequent_quadgram_cnt.append(ngram_counts.most_common(1)[0][1])
            else:
                most_frequent_pentagram.append(ngram_counts.most_common(1)[0][0])
                most_frequent_pentagram_cnt.append(ngram_counts.most_common(1)[0][1])        
```

```python
final_df = pd.DataFrame(list(zip(unique_mo_codes, num_records, most_frequent_unigram, most_frequent_unigram_cnt, most_frequent_bigram, most_frequent_bigram_cnt, most_frequent_trigram,most_frequent_trigram_cnt, most_frequent_quadgram, most_frequent_quadgram_cnt, most_frequent_pentagram,  most_frequent_pentagram_cnt)), columns = ['unique_mo_codes', 'num_records', 'most_frequent_unigram', 'most_frequent_unigram_cnt', 'most_frequent_bigram', 'most_frequent_bigram_cnt', 'most_frequent_trigram','most_frequent_trigram_cnt', 'most_frequent_quadgram', 'most_frequent_quadgram_cnt', 'most_frequent_pentagram',  'most_frequent_pentagram_cnt']) 
```
