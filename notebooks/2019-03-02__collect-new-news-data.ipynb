# Get Text

```python
from bs4 import BeautifulSoup
from newspaper import Article
from tqdm import tqdm_notebook as tqdm
from io import StringIO
import lxml
from lxml.html.clean import Cleaner

url = 'http://fox13now.com/2013/12/30/new-year-new-laws-obamacare-pot-guns-and-drones/'
url = 'https://www.nytimes.com/2020/03/02/us/politics/bernie-sanders-democratic-party.html'

article = Article(url)
article.download()
html = article.html
```

```python
cleaner = Cleaner()
cleaner.javascript = True # This is True because we want to activate the javascript filter
cleaner.style = True      # This is True because we want to activate the styles & stylesheet filter

def clean_me(html):
    clean_html = lxml.html.tostring(cleaner.clean_html(lxml.html.parse(StringIO(html))))
    soup = BeautifulSoup(clean_html)
    paragraphs = []
    for p in soup.find_all('p'):
        paragraphs.append(p.get_text())
    return paragraphs
```

# Get URLs from NYTimes

```python
htmls = []
for url in tqdm(test_df['web_url'][:200], total=200):
    article = Article(url)
    article.download()
    html = article.html
    htmls.append(html)
```

```python
cleaned_html = []
for h in htmls:
    cleaned_html.append(clean_me(h))
```
